As in the lecture, you will use the motivating example of housing price prediction.  
This lab will use a simple data set with only two data points - a house with 1000 square feet(sqft) sold for \\$300,000 and a house with 2000 square feet sold for \\$500,000. These two points will constitute our *data or training set*. In this lab, the units of size are 1000 sqft and the units of price are 1000s of dollars.

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

You would like to fit a linear regression model (shown above as the blue straight line) through these two points, so you can then predict price for other houses - say, a house with 1200 sqft.

```python
#import the necessary libraries:-
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dataset: {x,y}
x_train = np.array( [ 1.0, 2.0 ] )
y_train = np.array( [ 300, 500 ] )
```

```python
#training examples
m = x_train.shape[0]
print( "Number of Training Examples:", m )
```

Plotting the given Data:-

```python
plt.scatter( x_train, y_train, marker = 'x', c = 'blue')
plt.title( "House Pricing in Boston." )
plt.xlabel( "Sq ft in 1000s." )
plt.ylabel( "Prices in 1000s of USD." )
plt.show()
```

```python
#model equation:- f(w,b) = wT*x(i) + b

#Initialize weights and biases with random values:- 
w1 = 100
b1 = 200

w2 = 200
b2 = 100

def model_function( x, w, b):
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    return f_wb

#Test the function:-
y1 = model_function( x_train, w1, b1)
y2 = model_function( x_train, w2, b2)
print(y1)
print(y2)
```

```python
#Plot the regression line from the model on the given dataset:-
plt.figure(figsize=( 20, 10 )) 

plt.subplot( 1, 2, 1 )
plt.scatter( x_train, y_train, marker ='x', c = 'blue' )
plt.plot( x_train, y1, c = 'green', label = 'Prediction' )
plt.title( "House Pricing in Boston." )
plt.xlabel( "Sq ft in 1000s." )
plt.ylabel( "Prices in 1000s of USD." )

plt.subplot( 1, 2, 2 )
plt.scatter( x_train, y_train, marker ='x', c = 'blue' )
plt.plot( x_train, y2, c = 'red', label = 'Prediction' )
plt.title( "House Pricing in Boston." )
plt.xlabel( "Sq ft in 1000s." )
plt.ylabel( "Prices in 1000s of USD." )
plt.suptitle("Varying Weights and Biases - LinearRegression")

plt.show()
```
